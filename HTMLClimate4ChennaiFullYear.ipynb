```python
import urllib.request 
from pprint import pprint
from html_table_parser import HTMLTableParser
import pandas as pd
import csv
import string
headers={'User-Agent':'Whatever'}
def url_get_contents(url):
    req= urllib.request.Request(url=url,headers=headers)
    f=urllib.request.urlopen(req)
    return f.read()
```

```python
xhtml_list=[url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/january-1/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/february-2/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/march-3/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/april-4/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/may-5/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/june-6/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/july-7/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/august-8/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/september-9/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/october-10/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/november-11/').decode('utf-8'),url_get_contents('https://en.climate-data.org/asia/india/tamil-nadu/chennai-1003222/t/december-12/').decode('utf-8')]
df=pd.DataFrame(columns=['Month','Average Temperature','Maximum Temperature','Minimum Temperature','Rain/Precipitation'])
df.to_csv('ChennaiYearlyFinalClimate.csv')
frames=[df]
for i in range(len(xhtml_list)):
    p=HTMLTableParser()
    p.feed(xhtml_list[i])
    pprint(p.tables[3])
    print("\n\nPANDAS DATAFRAME\n")
    print(pd.DataFrame(p.tables[4]))
    months_list=[]
    average_temp_list=[]
    max_temp_list=[]
    min_temp_list=[]
    rain_list=[]
    
    for j in range(1,len(p.tables[4])):
        months_list.append(p.tables[4][j][0])
        average_temp_list.append(p.tables[4][j][1])
        max_temp_list.append(p.tables[4][j][2])
        min_temp_list.append(p.tables[4][j][3])
        rain_list.append(p.tables[4][j][4])
    dfi=pd.DataFrame(list(zip(months_list, average_temp_list,max_temp_list,min_temp_list,rain_list)))
    dfi.to_csv('ChennaiYearlyFinalClimate.csv', mode = 'a', header = False)


input_file = open('ChennaiYearlyFinalClimate.csv', 'r')
output_file = open('ChennaiYearlyFinalRevised.csv', 'w')
data = csv.reader(input_file)
writer = csv.writer(output_file)
specials = 'Â°'
for line in data:
    line = [value.replace(specials, '') for value in line]
    writer.writerow(line)
input_file.close()
output_file.close()
```
