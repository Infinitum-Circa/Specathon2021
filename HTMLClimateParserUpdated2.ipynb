```python
import urllib.request 
from pprint import pprint
from html_table_parser import HTMLTableParser
import pandas as pd
import csv
headers={'User-Agent':'Whatever'}
def url_get_contents(url):
    req= urllib.request.Request(url=url,headers=headers)
    f=urllib.request.urlopen(req)
    return f.read()
```

```python
xhtml=url_get_contents('https://en.climate-data.org/asia/india/hyderabad/hyderabad-2801/t/january-1/').decode('utf-8')
p=HTMLTableParser()
p.feed(xhtml)
pprint(p.tables[3])
print("\n\nPANDAS DATAFRAME\n") 
print(pd.DataFrame(p.tables[4]))
list1=[]
list2=[]
for i in range(len(p.tables[4])):
    list1.append(p.tables[4][i])
print(list1)
months_list=[]
average_temp_list=[]
max_temp_list=[]
min_temp_list=[]
rain_list=[]
for i in range(1,len(p.tables[4])):
    months_list.append(p.tables[4][i][0])
    average_temp_list.append(p.tables[4][i][1])
    max_temp_list.append(p.tables[4][i][2])
    min_temp_list.append(p.tables[4][i][3])
    rain_list.append(p.tables[4][i][4])
print(months_list)
print(average_temp_list)
print(max_temp_list)
print(min_temp_list)
print(rain_list)
df = pd.DataFrame(list(zip(months_list, average_temp_list,max_temp_list,min_temp_list,rain_list)), 
               columns =['Month', 'Average Temperature','Maximum Temperature','Minimum Temperature','Rain/Precipitation'])

print(df.head)
#df.to_csv('HyderabadJanuary.csv')
with open('HyderabadJanuaryClimate.csv','w',newline='') as csvfile:
        fieldnames=['Month','Average Temperature','Maximum Temperature','Minimum Temperature','Rain/Precipitation']
        thewriter=csv.DictWriter(csvfile,fieldnames=fieldnames)
        thewriter.writeheader()
        for i in range(0,len(p.tables[4])-1):
            thewriter.writerow({'Month':df.at[i,'Month'],'Average Temperature':df.at[i,'Average Temperature'],'Maximum Temperature':df.at[i,'Maximum Temperature'],'Minimum Temperature':df.at[i,'Minimum Temperature'],'Rain/Precipitation':df.at[i,'Rain/Precipitation']})
            

```
